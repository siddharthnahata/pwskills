1.  What is Deep Learning? Briefly describe how it evolved and how it differs 
from traditional machine learning.

Ans- Deep Learning is a subset of Machine Learning that uses artificial neural networks with multiple layers to automatically learn patterns and representations from large amounts of data. It mimics the way the human brain processes information by extracting features step by step.

It evolved from simple perceptrons in the 1950s–60s to advanced neural networks in the 1980s, and with the rise of big data and high computing power (GPUs) in the 2010s, deep learning achieved breakthroughs in image recognition, natural language processing, and speech recognition.

The difference from traditional Machine Learning is that ML often requires manual feature extraction and reaches a limit in performance with complex data, while Deep Learning automatically extracts features and improves with more data and computational power.

2. Explain the basic architecture and functioning of a Perceptron. What are its 
limitations? 

Ans-
A Perceptron is the simplest type of artificial neural network, introduced by Frank Rosenblatt in 1958. Its architecture consists of:

* Inputs (x₁, x₂, …, xn)

* Weights (w₁, w₂, …, wn) assigned to each input

* Summation function that computes the weighted sum (Σ wi·xi + b)

An Activation function (usually step function) that outputs either 0 or 1 depending on whether the sum crosses a threshold.

Functioning:
The perceptron takes input values, multiplies them by their weights, adds a bias, and passes the result through the activation function to produce an output (used for binary classification). During training, weights are updated to minimize errors.

Limitations:

* Can only solve linearly separable problems (e.g., cannot solve XOR).

* Limited learning capacity due to single-layer structure.

* Step activation makes it non-differentiable (not suitable for gradient-based learning).

* Cannot handle complex patterns or non-linear data.

3.  Describe the purpose of activation function in neural networks. Compare 
Sigmoid, ReLU, and Tanh functions. 

Ans-
The purpose of an activation function is to introduce non-linearity into a neural network, enabling it to learn complex patterns. Without activation functions, the network would behave like a simple linear model regardless of the number of layers.

Comparison of Activation Functions:

1. Sigmoid Function (σ(x) = 1 / (1 + e^(-x))):

* Output range: (0, 1)

* Good for probabilities.

* Limitations: Causes vanishing gradient problem, slow convergence.

2. Tanh Function (tanh(x)):

* Output range: (-1, 1)

* Zero-centered, performs better than Sigmoid.

* Limitations: Still suffers from vanishing gradients.

3. ReLU (Rectified Linear Unit, f(x) = max(0, x)):

* Output range: [0, ∞)

* Computationally efficient, helps avoid vanishing gradients.

* Limitations: Can cause dying ReLU problem (neurons stuck at 0).

Summary:

* Sigmoid → good for probabilities but slow.

* Tanh → better than Sigmoid, still limited.

* ReLU → most widely used due to efficiency and performance.

5. What is the role of optimizers in neural networks? Compare Gradient 
Descent, Adam, and RMSprop.  

Ans- 

Optimizers update weights to minimize the loss function. They control **how fast and stable** a network learns.

---

### Gradient Descent
- **Rule:** \(\theta = \theta - \eta \nabla J(\theta)\)  
- **Pros:** Simple, well-known.  
- **Cons:** Slow, sensitive to learning rate.  

---

### RMSprop
- **Idea:** Scales learning rate by moving avg. of squared gradients.  
- **Pros:** Works well for non-stationary data, RNNs.  
- **Cons:** Sensitive to decay rate.  

---

### Adam
- **Idea:** Combines **Momentum + RMSprop**.  
- **Pros:** Fast, widely used, good for sparse gradients.  
- **Cons:** May generalize worse than SGD, more hyperparams.  

---

### Quick Comparison

| Optimizer        | Key Idea                  | Pros             | Cons                    |
|------------------|--------------------------|------------------|-------------------------|
| Gradient Descent | Fixed LR, raw gradients  | Simple           | Slow, unstable          |
| RMSprop          | Squared grad scaling     | Good for RNNs    | Needs tuning            |
| Adam             | Momentum + RMSprop       | Fast, effective  | May overfit, many params |

6. Write a Python program to implement a single-layer perceptron from 
scratch using NumPy to solve the logical AND gate

```python
# AND gate dataset
X = np.array([[0,0], [0,1], [1,0], [1,1]])   # inputs
y = np.array([0, 0, 0, 1])                   # outputs

# Parameters
epochs = 20
lr = 0.1   # learning rate

# Initialize weights and bias
weights = np.zeros(X.shape[1])
bias = 0

# Activation function (step function)
def step_function(x):
    return np.where(x >= 0, 1, 0)

# Training
for epoch in range(epochs):
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        y_pred = step_function(linear_output)
        
        # Perceptron update rule
        error = y[i] - y_pred
        weights += lr * error * X[i]
        bias += lr * error

    print(f"Epoch {epoch+1}: Weights={weights}, Bias={bias}")

# Testing
print("\nTesting AND gate:")
for i in range(len(X)):
    linear_output = np.dot(X[i], weights) + bias
    y_pred = step_function(linear_output)
    print(f"Input: {X[i]} → Predicted: {y_pred}, Actual: {y[i]}")
```

7. Implement and visualize Sigmoid, ReLU, and Tanh activation functions 
using Matplotlib.

```python
# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Input range
x = np.linspace(-10, 10, 500)

# Compute activations
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# Plot
plt.figure(figsize=(12, 6))

# Sigmoid
plt.subplot(1, 3, 1)
plt.plot(x, y_sigmoid, color="blue")
plt.title("Sigmoid Activation")
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# ReLU
plt.subplot(1, 3, 2)
plt.plot(x, y_relu, color="green")
plt.title("ReLU Activation")
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Tanh
plt.subplot(1, 3, 3)
plt.plot(x, y_tanh, color="red")
plt.title("Tanh Activation")
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.tight_layout()
plt.show()
```

8. Visualize the loss and accuracy curves for a neural network model trained 
on the Fashion MNIST dataset. Interpret the training behavior.

```python
import tensorflow as tf
from tensorflow.keras import layers, models

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

# Normalize the data
x_train, x_test = x_train / 255.0, x_test / 255.0

# Build a simple neural network model
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(x_train, y_train, epochs=10, batch_size=128, 
                    validation_data=(x_test, y_test), verbose=0)


# Plot loss and accuracy curves
plt.figure(figsize=(12, 5))

# Loss curves
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)

# Accuracy curves
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title("Accuracy Curve")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

history.history  # Return history data for interpretation
```

## Fraudulent Transaction Detection - Deep Learning Workflow

### Problem Statement
We need to detect fraudulent transactions in a large, imbalanced dataset with structured features (transaction amount, merchant ID, customer location, etc.).

---

### 1. Model Design
- I would use a **Multilayer Perceptron (MLP)** since we are dealing with **structured tabular data**.  
- Input layer → hidden dense layers with non-linear activations → output layer with sigmoid for binary classification.

---

### 2. Activation & Loss Function
- **Hidden Layers**: ReLU (faster convergence, avoids vanishing gradient).  
- **Output Layer**: Sigmoid (gives probability between 0 and 1).  
- **Loss Function**: Binary Cross-Entropy, because it’s a binary classification problem.  
  \[
  L = -\frac{1}{N}\sum [y \log(p) + (1-y)\log(1-p)]
  \]

---

### 3. Training & Evaluation (Handling Imbalance)
- The dataset is **imbalanced** (more legitimate transactions than fraud).  
- Strategies:  
  - Use **class weights** in training (`class_weight` in Keras).  
  - Use **precision, recall, F1-score, AUC-ROC** instead of only accuracy.  
  - Possible **SMOTE/oversampling** or **undersampling** before training.  

---

### 4. Optimizer & Preventing Overfitting
- **Optimizer**: Adam (fast, adaptive learning rate).  
- **Regularization**: Dropout, L2 weight decay, Early stopping.  
- **Validation set** to monitor generalization.  

---


```python
# CODE CELL 1: Imports
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.datasets import make_classification
from tensorflow.keras import layers, models
from sklearn.utils.class_weight import compute_class_weight
```

```python
# CODE CELL 2: Synthetic Data Generator
def generate_fraud_data(n_samples=10000, imbalance_ratio=0.05, n_features=5):
    """
    Generate synthetic fraud detection dataset.
    
    n_samples: total samples
    imbalance_ratio: fraction of fraud cases (default 5%)
    n_features: number of features
    
    Returns: X (features), y (labels)
    """
    X, y = make_classification(n_samples=n_samples,
                               n_features=n_features,
                               n_informative=10,
                               n_redundant=5,
                               n_classes=2,
                               weights=[1-imbalance_ratio, imbalance_ratio],
                               flip_y=0.01,
                               random_state=42)
    return X, y

# Generate dataset
X, y = generate_fraud_data(n_samples=20000, imbalance_ratio=0.03, n_features=30)
print("Dataset shape:", X.shape, y.shape)
print("Fraud cases:", np.sum(y))
```

```python
# CODE CELL 3: Preprocessing
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Scale features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# CODE CELL 4: Model Definition
model = models.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    layers.Dropout(0.3),
    layers.Dense(32, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(1, activation='sigmoid')  # binary classification
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# CODE CELL 5: Handle Class Imbalance
class_weights = compute_class_weight(class_weight='balanced',
                                     classes=np.unique(y_train),
                                     y=y_train)
class_weights = dict(enumerate(class_weights))
print("Class Weights:", class_weights)
```

```python
# CODE CELL 6: Train Model
history = model.fit(X_train, y_train,
                    validation_split=0.2,
                    epochs=15,
                    batch_size=256,
                    class_weight=class_weights,
                    verbose=1)
```

```python
# CODE CELL 7: Evaluation
y_pred = (model.predict(X_test) > 0.5).astype(int)
print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, model.predict(X_test)))
```

```python
# CODE CELL 8: Plot Loss & Accuracy
plt.figure(figsize=(12, 5))

# Loss curve
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label="Train Loss")
plt.plot(history.history['val_loss'], label="Validation Loss")
plt.title("Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)

# Accuracy curve
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label="Train Accuracy")
plt.plot(history.history['val_accuracy'], label="Validation Accuracy")
plt.title("Accuracy Curve")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)

plt.show()
```

## Interpretation
- **Training & Validation Loss** should decrease over epochs.  
- **Accuracy** rises but since the dataset is imbalanced, **ROC-AUC** and **Recall** are more important.  
- With class weights, the model should learn to detect fraud cases better.
