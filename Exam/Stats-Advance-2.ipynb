3. What is the significance level in hypothesis testing, and why is it important.

<b>Ans-</b> In hypothesis testing, the significance level (denoted by Î±, alpha) is the probability of rejecting the null hypothesis when it is actually true. It represents the researcher's willingness to accept a Type I error, which is falsely rejecting a true null hypothesis. It's a crucial threshold for determining statistical significance, helping researchers decide whether observed results are likely due to chance or a real effect. 

4.  What does a P-value represent in hypothesis testing

5. How do you interpret the P-value in hypothesis testing

<b>Ans-</b> The p-value only tells you how likely the data you have observed is to have occurred under the null hypothesis. If the p-value is below your threshold of significance (typically p < 0.05), then you can reject the null hypothesis, but this does not necessarily mean that your alternative hypothesis is true.

9.  How do you calculate the Z-score, and what does it represent in hypothesis testing

<b>Ans-</b> Lets take any example. and Z-Score = (x_bar - mu) / (p_sigma / n ** 0.5)

A tire company claims that its tires last at least 60,000 km on average.

An independent lab thinks the average is actually less than that.

They test a sample of 49 tires and find:

Sample mean = 58,900 km

Standard deviation = 2,100 km

Significance level = 0.01

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

population_mean = 60000
sample_mean = 58900
n = 49
population_sigma = 2100
alpha = 0.01

z = (sample_mean - population_mean) / (population_sigma / (n ** 0.5))
z_critical = norm.ppf(alpha)

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')

x_crit = np.linspace(-4, z_critical, 100)
y_crit = norm.pdf(x_crit)
plt.fill_between(x_crit, y_crit, color='red', alpha=0.4, label='Critical Region (Î± = 0.01)')

plt.axvline(z, color='black', linestyle='--', label=f'Z-stat = {z:.2f}')
plt.axvline(z_critical, color='red', linestyle='--', label=f'Z-critical = {z_critical:.2f}')

plt.title('Z-Test: One-Tailed Test (Left Tail)')
plt.xlabel('Z value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

18. What is the Chi-square goodness of fit test, and how is it applied

<b>Ans-</b> It is a one tail test as numbers are squared here we compare expected with observed and try to see or check is difference to big or it is just random. It is applied with a formual sigma (Obeserve - Expected)^2/Expected. Then for alpha we get critical point from table and check our chi stats is greater or not if greater then we reject it if lesser the we fail to reject.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2

data_points = ['1', '2', '3', '4', '5', '6']
observed = np.array([12, 9, 11, 13, 8, 7])
expected = np.array([10, 10, 10, 10, 10, 10])
df = len(data_points) - 1

chi_square_stat = np.sum((observed - expected) ** 2 / expected)

critical_value = chi2.ppf(0.95, df)

x = np.linspace(0, 20, 300)
y = chi2.pdf(x, df)

plt.plot(x, y, label='Chi-Squared Distribution', color='blue')
plt.axvline(chi_square_stat, color='black', linestyle='--', label=f'Chi-Square Stat = {chi_square_stat:.2f}')
plt.axvline(critical_value, color='red', linestyle='--', label=f'Critical Value = {critical_value:.2f}')
plt.fill_between(x, y, where=(x >= critical_value), color='red', alpha=0.4, label='Critical Region (Î± = 0.05)')

plt.title('Chi-Square Test: Observed vs Critical Value')
plt.xlabel('Chi-Square Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(mu0, mu1, sigma, n, alpha=0.05, alternative='two-sided'):

    # Standard error
    se = sigma / np.sqrt(n)

    # Critical z-values
    if alternative == 'two-sided':
        z_crit = norm.ppf(1 - alpha/2)
        z_left = -z_crit
        z_right = z_crit
    elif alternative == 'left':
        z_right = norm.ppf(alpha)
    elif alternative == 'right':
        z_right = norm.ppf(1 - alpha)

    # X ranges
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # Distributions under H0 and H1
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label='Hâ Distribution', color='blue')
    plt.plot(x, y1, label='Hâ Distribution', color='green')

    # Type I Error (red area)
    if alternative == 'two-sided':
        plt.fill_between(x, y0, where=(x <= mu0 + z_left * se), color='red', alpha=0.4, label='Type I Error (Î±)')
        plt.fill_between(x, y0, where=(x >= mu0 + z_right * se), color='red', alpha=0.4)
    elif alternative == 'right':
        plt.fill_between(x, y0, where=(x >= mu0 + z_right * se), color='red', alpha=0.4, label='Type I Error (Î±)')
    elif alternative == 'left':
        plt.fill_between(x, y0, where=(x <= mu0 + z_right * se), color='red', alpha=0.4, label='Type I Error (Î±)')

    # Type II Error (blue area under Hâ)
    if alternative == 'two-sided':
        beta = norm.cdf(mu0 + z_right * se, mu1, se) - norm.cdf(mu0 + z_left * se, mu1, se)
        plt.fill_between(x, y1, where=((x >= mu0 + z_left * se) & (x <= mu0 + z_right * se)), color='purple', alpha=0.3, label='Type II Error (Î²)')
    elif alternative == 'right':
        beta = norm.cdf(mu0 + z_right * se, mu1, se)
        plt.fill_between(x, y1, where=(x <= mu0 + z_right * se), color='purple', alpha=0.3, label='Type II Error (Î²)')
    elif alternative == 'left':
        beta = 1 - norm.cdf(mu0 + z_right * se, mu1, se)
        plt.fill_between(x, y1, where=(x >= mu0 + z_right * se), color='purple', alpha=0.3, label='Type II Error (Î²)')

    plt.title("Type I (Î±) and Type II (Î²) Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"â Type II Error (Î²): {beta:.4f}")
    print(f"â Power of the test (1 - Î²): {1 - beta:.4f}")

visualize_type1_type2(70, 75, 2, 64, 0.05)
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
def bayesian_inference():
    # Given:
    P_D = 0.01           
    P_not_D = 1 - P_D     # Probability of no disease

    P_Pos_given_D = 0.99  # True positive rate
    P_Pos_given_not_D = 0.05  # False positive rate

    # Total probability of testing positive
    P_Pos = (P_Pos_given_D * P_D) + (P_Pos_given_not_D * P_not_D)

    # Apply Bayes' Theorem
    P_D_given_Pos = (P_Pos_given_D * P_D) / P_Pos

    print(f"Probability you actually have the disease given a positive test: {P_D_given_Pos:.4f} ({P_D_given_Pos*100:.2f}%)")

bayesian_inference()
```

12. Perform a Chi-square test for independence between two categorical variables in Python

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

data = pd.DataFrame({
    'Tea':   [20, 30, 10],   
    'Coffee': [15, 25, 20]
}, index=['Young', 'Adult', 'Senior'])

print("Contingency Table:\n")
print(data)

# Perform Chi-Square Test
chi2_stat, p_value, dof, expected = chi2_contingency(data)

print("\nChi-Square Statistic:", round(chi2_stat, 4))
print("Degrees of Freedom:", dof)
print("P-Value:", round(p_value, 4))
print("\nExpected Frequencies:\n", pd.DataFrame(expected, index=data.index, columns=data.columns))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nResult: Reject Hâ â There IS a significant relationship between Age and Drink Preference.")
else:
    print("\nResult: Fail to reject Hâ â No significant relationship between Age and Drink Preference.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2

data_points = ['1', '2', '3', '4', '5', '6']
observed = np.array([16, 5, 11, 16, 5, 7])
expected = np.array([10, 10, 10, 10, 10, 10])
df = len(data_points) - 1
alpha = 0.05


chi2_stats = sum((observed - expected)**2/expected)
chi2_critical = chi2.ppf(1 - alpha, df)
x = np.linspace(0, 15, 500)
y = chi2.pdf(x, df)

print(f"Chi-Square Stats: {chi2_stats.round(2)}")
print(f"Chi-Square Critical: {chi2_critical.round(2)}")

if chi2_stats > chi2_critical:
    print("--> Reject the null hypothesis")
else:
    print("--> Fail to Reject the null hypothesis")
```

16. Implement an F-test using Python to compare the variances of two random samples

```python
import numpy as np
from scipy.stats import f

np.random.seed(0)
sample1 = np.random.normal(loc=10, scale=3, size=30)
sample2 = np.random.normal(loc=10, scale=2, size=25)

var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

n1, n2 = len(sample1), len(sample2)
df1, df2 = n1 - 1, n2 - 1

if var1 > var2:
    F = var1 / var2
    df_num, df_den = df1, df2
else:
    F = var2 / var1
    df_num, df_den = df2, df1

p_value = 2 * min(f.cdf(F, df_num, df_den), 1 - f.cdf(F, df_num, df_den))

print(f"Variance 1: {var1:.4f}")
print(f"Variance 2: {var2:.4f}")
print(f"F-statistic: {F:.4f}")
print(f"Degrees of freedom: ({df_num}, {df_den})")
print(f"P-value (two-tailed): {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("--> Reject Hâ: Variances are significantly different.")
else:
    print("--> Fail to reject Hâ: No significant difference in variances.")
```

19.  Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

```python
import numpy as np
import pandas as pd
from scipy.stats import shapiro, levene, f_oneway
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0)
group_A = np.random.normal(loc=70, scale=5, size=30)
group_B = np.random.normal(loc=72, scale=5, size=30)
group_C = np.random.normal(loc=74, scale=5, size=30)

groups = [group_A, group_B, group_C]
group_labels = ['A']*30 + ['B']*30 + ['C']*30
all_data = np.concatenate(groups)


print("=== Checking Assumptions ===")

for i, group in enumerate(groups):
    stat, p = shapiro(group)
    print(f"Group {chr(65+i)} - Shapiro-Wilk p = {p:.4f} â", "Normal" if p > 0.05 else "Not Normal")

stat, p = levene(*groups)
print(f"\nLeveneâs Test p = {p:.4f} â", "Equal Variance" if p > 0.05 else "Unequal Variance")

print("\nIndependence: Assumed based on study design.")

f_stat, p_value = f_oneway(*groups)
print(f"\n=== One-Way ANOVA ===")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Reject Hâ â At least one group mean is significantly different.")
else:
    print("Fail to reject Hâ â No significant difference between group means.")

df = pd.DataFrame({'Score': all_data, 'Group': group_labels})
plt.figure(figsize=(8, 5))
sns.boxplot(x='Group', y='Score', data=df)
plt.title('One-Way ANOVA: Group Score Comparison')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the
results

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

np.random.seed(1)
data = pd.DataFrame({
    'Score': np.random.normal(70, 5, 60),
    'Method': ['Online']*20 + ['Offline']*20 + ['Hybrid']*20,
    'Gender': (['Male', 'Female']*10)*3
})

# Add variation by method and gender
data.loc[data['Method'] == 'Offline', 'Score'] += 3
data.loc[data['Gender'] == 'Female', 'Score'] += 2
data.loc[(data['Method'] == 'Hybrid') & (data['Gender'] == 'Female'), 'Score'] += 2

#Two-Way ANOVA using statsmodels
model = ols('Score ~ C(Method) + C(Gender) + C(Method):C(Gender)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("=== Two-Way ANOVA Results ===\n")
print(anova_table)

# Step 3: Visualization - Interaction Plot
plt.figure(figsize=(8, 5))
sns.pointplot(data=data, x='Method', y='Score', hue='Gender', dodge=True, markers=['o', 's'], capsize=0.1   )
plt.title('Interaction Plot: Method vs Gender on Score')
plt.grid(True, linestyle='--', alpha=0.5)
plt.ylabel("Mean Score")
plt.show()
```

21. Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing


A fitness coach wants to compare the effectiveness of three workout programs (Program A, B, and C) on weight loss (in kg) over 4 weeks. The data collected is:

Program A: [2.5, 2.9, 2.7]	Program B: [3.0, 2.8, 3.2]	Program C: [4.2, 3.9, 4.0]

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f, f_oneway

a = [2.5, 2.9, 2.7]
b = [3.0, 2.8, 3.2]
c = [4.2, 3.9, 4.0]

groups = [a, b, c]

k = len(groups)
N = sum(len(group) for group in groups)
df1 = k - 1
df2 = N - k

f_stat, p_value = f_oneway(*groups)
alpha = 0.05

f_critical = f.ppf(1 - alpha, dfn=df1, dfd=df2)


x = np.linspace(0, 40, 50)
y = f.pdf(x, dfn=df1, dfd=df2)


plt.figure(figsize=(10, 6))
plt.plot(x, y, label='F-distribution', color='blue')
plt.axvline(f_stat, color="black", linestyle="--", label=f'F-stat = {f_stat:.2f}')
plt.axvline(f_critical, color="red", linestyle="--", label=f'F-stat = {f_critical:.2f}')

plt.legend()
plt.show()
```

24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

```python
import numpy as np
from scipy.stats import chi2


np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=25) 

sigma_0_squared = 25 

n = len(data)
sample_var = np.var(data, ddof=1)
chi2_stat = (n - 1) * sample_var / sigma_0_squared

alpha = 0.05
chi2_critical_low = chi2.ppf(alpha / 2, df=n - 1)
chi2_critical_high = chi2.ppf(1 - alpha / 2, df=n - 1)

print("=== Chi-Square Test for Variance ===")
print(f"Sample Variance: {sample_var:.4f}")
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"Critical Region: < {chi2_critical_low:.4f} or > {chi2_critical_high:.4f}")

if chi2_stat < chi2_critical_low or chi2_stat > chi2_critical_high:
    print("Reject Hâ â Population variance is significantly different.")
else:
    print("Fail to reject Hâ â No significant difference in variance.")
```

26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the resultsD

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

np.random.seed(42)
group1 = np.random.normal(loc=50, scale=5, size=30)   
group2 = np.random.normal(loc=50, scale=8, size=30)

var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)

if var1 > var2:
    F = var1 / var2
    df1, df2 = len(group1) - 1, len(group2) - 1
else:
    F = var2 / var1
    df1, df2 = len(group2) - 1, len(group1) - 1

alpha = 0.05
f_critical_low = f.ppf(alpha / 2, df1, df2)
f_critical_high = f.ppf(1 - alpha / 2, df1, df2)
p_value = 2 * min(f.cdf(F, df1, df2), 1 - f.cdf(F, df1, df2))

# Step 5: Print results
print("=== F-Test for Variance ===")
print(f"Variance 1: {var1:.4f}")
print(f"Variance 2: {var2:.4f}")
print(f"F-statistic: {F:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Critical Region: F < {f_critical_low:.4f} or F > {f_critical_high:.4f}")

if F < f_critical_low or F > f_critical_high:
    print("Reject Hâ: Variances are significantly different.")
else:
    print("Fail to reject Hâ: No significant difference in variances.")

x = np.linspace(0.1, 5, 500)
y = f.pdf(x, df1, df2)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='F-distribution', color='blue')
plt.axvline(F, color='black', linestyle='--', label=f'F = {F:.2f}')
plt.axvline(f_critical_low, color='red', linestyle='--', label='Lower Critical Value')
plt.axvline(f_critical_high, color='red', linestyle='--', label='Upper Critical Value')
plt.fill_between(x, y, where=(x < f_critical_low) | (x > f_critical_high), color='red', alpha=0.3, label='Rejection Region')
plt.title('F-Test: Comparing Two Variances')
plt.xlabel('F Value')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```
