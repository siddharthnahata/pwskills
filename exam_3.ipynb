6. Explain the concept of generators in Python and how they are defined

Ans- A generator in Python is defined using def and the yield keyword. It returns values one at a time, pausing after each yield and resuming when needed. This makes it memory-efficient and useful for large or infinite sequences. You can use it with loops or next() to get values.

```python
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

gen = count_up_to(3)
print(next(gen))
print(next(gen))
print(next(gen))
```

the list

```python
nums_list = list(range(21))

def cal_even(n):
    sum = 0
    for i in n:
        if i % 2 == 0:
            sum += i
    return sum    
cal_even(nums_list)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200

```python
nums = list(range(1,201))

prime_nums = []

def is_prime(n):
    if n <= 1:
        return False
    elif n == 2:
        return True
    elif n % 2 == 0:
        return False
    else:
        for i in range(3, int(n**0.5) + 1, 2):
            if n % i == 0:
                return False
            return True

for num in nums:
    if is_prime(num): 
        prime_nums.append(num)
print(prime_nums)
```

5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
def fib(max):
    a, b = 0, 1
    for _ in range(max):
        yield a
        a, b = b, a + b
for num in fib(10):
    print(num, end=" ")
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit

```python
celsius = [5, 12, 28, 32, 45]

def cel_to_fah(cel):
    fah = (9 / 5 * cel) + 32
    return fah
fahrenheit = list(map(cel_to_fah, celsius))
print(fahrenheit)
```
