1. What is a Convolutional Neural Network (CNN), and how does it differ from
traditional fully connected neural networks in terms of architecture and performance on
image data?

Ans- A Convolutional Neural Network (CNN) is a type of deep learning model mainly used for image data. It works by applying small filters, also called kernels, over the image to automatically extract important spatial features such as edges, shapes, and textures. Unlike traditional fully connected neural networks, where every neuron is connected to all inputs and the image has to be flattened into a single vector (losing spatial information), CNNs preserve the 2D structure of the image. They use convolution and pooling layers to detect patterns and reduce the number of parameters, making them more efficient and accurate for visual tasks. In short, CNNs learn to identify features directly from images, while fully connected networks simply compute numerical relationships without understanding spatial structure.

2. Discuss the architecture of LeNet-5 and explain how it laid the foundation
for modern deep learning models in computer vision. Include references to its original
research paper.

Ans- LeNet-5, is one of the earliest Convolutional Neural Networks designed for handwritten digit recognition. It consists of seven layers, including convolutional, pooling, and fully connected layers. The network takes a 32×32 grayscale image as input and gradually extracts features through convolution and subsampling layers before classification in the final dense layers. LeNet-5 introduced key concepts like convolution, pooling, and hierarchical feature extraction, which became the foundation for modern deep learning models such as AlexNet, VGG, and ResNet.

3.  Compare and contrast AlexNet and VGGNet in terms of design principles,
number of parameters, and performance. Highlight key innovations and limitations of
each.

Ans- AlexNet was the first deep CNN to achieve major success on ImageNet. It had 8 layers and around 60 million parameters, using innovations like ReLU activation, dropout, and GPU training. However, it used large filters and was quite memory-intensive. VGGNet (Simonyan & Zisserman, 2014) improved upon AlexNet with a simpler and deeper design, using small 3×3 filters stacked in depth. It achieved better feature extraction but had about 138 million parameters, making it computationally heavy. In short, AlexNet started the deep learning revolution, while VGGNet refined it with a cleaner and deeper architecture.

4. What is transfer learning in the context of image classification? Explain
how it helps in reducing computational costs and improving model performance with
limited data.

Ans- Transfer learning in image classification is a technique where a model pre-trained on a large dataset (like ImageNet) is reused for a new, smaller task. Instead of training a CNN from scratch, the pre-trained model’s learned features—such as edges, textures, and shapes—are retained, and only the final layers are fine-tuned for the new dataset. This approach greatly reduces computational cost because most of the feature extraction is already learned, and it also improves performance when limited data is available, as the model starts with strong, general visual features rather than random weights.

5.  Describe the role of residual connections in ResNet architecture. How do
they address the vanishing gradient problem in deep CNNs?  

Ans- In the ResNet (Residual Network) architecture, residual connections or skip connections allow the input of a layer to bypass one or more layers and be added directly to the output. This means the network learns a residual mapping (the difference between input and output) instead of the full transformation. These connections help prevent the vanishing gradient problem, which often occurs in very deep networks, by allowing gradients to flow directly through the skip paths during backpropagation. As a result, ResNet can be trained with hundreds of layers efficiently, improving both convergence speed and accuracy without degradation in performance.

6.  Implement the LeNet-5 architectures using Tensorflow or PyTorch to
classify the MNIST dataset. Report the accuracy and training time.

```python
from tensorflow import keras
from keras.datasets import mnist
from tensorflow.keras.layers import Conv2D, MaxPooling2D, AveragePooling2D
from keras.layers import Dense, Flatten
from keras.models import Sequential
import time

(X_train, y_train), (X_test, y_test) = mnist.load_data()

# perfroming reshaping
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1)

print(f"Number of sample in Training data: {X_train.shape[0]}")
print(f"Number of sample in Testing data: {X_test.shape[0]}")

# normalizing the values
X_train = X_train / 255
X_test = X_test / 255

# one hot encoding the y
y_train = keras.utils.to_categorical(y_train)
y_test = keras.utils.to_categorical(y_test)

# neural newtork architecture for lenet 6
model = Sequential([
    Conv2D(6, kernel_size=(5, 5), activation='relu', input_shape=(28, 28, 1)), # 24 x 24 x 6
    MaxPooling2D(pool_size=(2,2)), # 12 x 12 x 6
    Conv2D(16, kernel_size=(5, 5), activation='relu'), # 8 x 8 x 16
    MaxPooling2D(pool_size=(2,2)), # 4 x 4 x 6
    Conv2D(120, kernel_size=(5, 5), activation='relu', padding='same'), # 4 x 4 x 120
    Flatten(),
    Dense(84, activation='relu'),
    Dense(10, activation='softmax')
])

print("Model Summary")
model.summary()

model.compile(loss=keras.metrics.categorical_crossentropy, optimizer=keras.optimizers.Adam(), metrics=['accuracy'])
# Record the start time
start_time = time.perf_counter()
model.fit(X_train, y_train, batch_size=128, epochs=20, verbose=1, validation_data=(X_test, y_test))
end_time = time.perf_counter()

score = model.evaluate(X_test, y_test)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
print(f"Training time: {end_time - start_time:.2f}s")
```

8. : Write a program to visualize the filters and feature maps of the first
convolutional layer of AlexNet on an example input image.

```python
import torch
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
import requests
from io import BytesIO

# ---- Load pretrained AlexNet ----
alexnet = models.alexnet(weights=models.AlexNet_Weights.IMAGENET1K_V1)
alexnet.eval()

# ---- Download a stable test image ----
url = "https://raw.githubusercontent.com/pytorch/hub/master/images/dog.jpg"  # ✅ works perfectly
response = requests.get(url)
image = Image.open(BytesIO(response.content)).convert("RGB")

# ---- Preprocess the image ----
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

img_tensor = transform(image).unsqueeze(0)

# ---- Get first convolutional layer ----
first_layer = alexnet.features[0]

# ---- Visualize filters ----
filters = first_layer.weight.data
plt.figure(figsize=(8, 8))
for i in range(16):  # show first 16 filters
    f = (filters[i] - filters[i].min()) / (filters[i].max() - filters[i].min())
    plt.subplot(4, 4, i + 1)
    plt.imshow(f.permute(1, 2, 0))
    plt.axis('off')
plt.suptitle("AlexNet - First Conv Layer Filters")
plt.show()

# ---- Visualize feature maps ----
with torch.no_grad():
    feature_maps = first_layer(img_tensor).squeeze(0)

plt.figure(figsize=(8, 8))
for i in range(16):  # show first 16 feature maps
    fmap = feature_maps[i]
    fmap = (fmap - fmap.min()) / (fmap.max() - fmap.min())
    plt.subplot(4, 4, i + 1)
    plt.imshow(fmap, cmap='gray')
    plt.axis('off')
plt.suptitle("AlexNet - Feature Maps")
plt.show()
```

9. Train a GoogLeNet (Inception v1) or its variant using a standard dataset
like CIFAR-10. Plot the training and validation accuracy over epochs and analyze
overfitting or underfitting.

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torchvision.models import googlenet
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

transform_train = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomCrop(32, padding=4),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5),
                         (0.5, 0.5, 0.5))
])
transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5),
                         (0.5, 0.5, 0.5))
])

trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform_train)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=128,
                                          shuffle=True, num_workers=2)
testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform_test)
testloader = torch.utils.data.DataLoader(testset, batch_size=128,
                                         shuffle=False, num_workers=2)
model = googlenet(weights=None, num_classes=10, aux_logits=True)
model = model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

epochs = 10
train_acc_list, val_acc_list = [], []

for epoch in range(epochs):
    model.train()
    correct, total = 0, 0
    for images, labels in trainloader:
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()

        # GoogLeNet returns multiple outputs
        outputs = model(images)
        if isinstance(outputs, torch.nn.modules.container.Sequential) or isinstance(outputs, tuple):
            main_out, aux2, aux1 = outputs
            loss = (criterion(main_out, labels) +
                    0.3 * criterion(aux1, labels) +
                    0.3 * criterion(aux2, labels))
        else:
            loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

        # accuracy on training batch
        _, predicted = torch.max(outputs.logits if hasattr(outputs, 'logits') else outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    train_acc = 100 * correct / total
    train_acc_list.append(train_acc)

    # Validation
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for images, labels in testloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            out = outputs.logits if hasattr(outputs, 'logits') else outputs
            _, predicted = torch.max(out.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    val_acc = 100 * correct / total
    val_acc_list.append(val_acc)

    print(f"Epoch [{epoch+1}/{epochs}] | Train Acc: {train_acc:.2f}% | Val Acc: {val_acc:.2f}%")

print("Training Complete")

plt.figure(figsize=(8,5))
plt.plot(range(1, epochs+1), train_acc_list, label='Train Accuracy')
plt.plot(range(1, epochs+1), val_acc_list, label='Validation Accuracy')
plt.title("GoogLeNet (Inception v1) on CIFAR-10")
plt.xlabel("Epochs")
plt.ylabel("Accuracy (%)")
plt.legend()
plt.grid(True)
plt.show()
```

## Step 1 - Import Libraries & Configuration

```python
import tensorflow as tf
from tensorflow.keras import layers, models, applications, optimizers
import matplotlib.pyplot as plt
```

## Step 2 – Data Loading & Augmentation

```python
def load_data(data_dir, img_size=224, batch_size=32):
    datagen = tf.keras.preprocessing.image.ImageDataGenerator(
        rescale=1./255,
        validation_split=0.2,
        horizontal_flip=True,
        zoom_range=0.2
    )
    train_ds = datagen.flow_from_directory(
        data_dir, target_size=(img_size, img_size),
        batch_size=batch_size, subset='training', class_mode='categorical'
    )
    val_ds = datagen.flow_from_directory(
        data_dir, target_size=(img_size, img_size),
        batch_size=batch_size, subset='validation', class_mode='categorical'
    )
    return train_ds, val_ds
```

## Step 3 – Build Transfer Learning Model

```python
def build_model(num_classes=3, img_size=224, lr=1e-4):
    base = applications.ResNet50(weights='imagenet', include_top=False, input_shape=(img_size, img_size, 3))
    base.trainable = False  # freeze base layers
    model = models.Sequential([
        base,
        layers.GlobalAveragePooling2D(),
        layers.Dropout(0.3),
        layers.Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer=optimizers.Adam(lr), loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

## Step 4 – Train & Fine-Tune

```python
def train_model(model, train_ds, val_ds, epochs=10):
    hist = model.fit(train_ds, validation_data=val_ds, epochs=epochs)
    return hist
```

## Step 5 – Plot Accuracy and Detect Overfitting

```python
def plot_history(hist):
    plt.plot(hist.history['accuracy'], label='Train Acc')
    plt.plot(hist.history['val_accuracy'], label='Val Acc')
    plt.title("Training vs Validation Accuracy")
    plt.xlabel("Epochs"); plt.ylabel("Accuracy")
    plt.legend(); plt.grid(); plt.show()

    train_acc = hist.history['accuracy'][-1]
    val_acc = hist.history['val_accuracy'][-1]
    gap = train_acc - val_acc
    print(f"Train: {train_acc:.2f} | Val: {val_acc:.2f}")
    print("Overfitting" if gap > 0.1 else "Underfitting" if (train_acc < 0.8 and val_acc < 0.8) else "Well fitted")
```
